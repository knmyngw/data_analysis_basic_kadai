```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# tight_layout -> レイアウト自動調整(推奨)
fig, axes = plt.subplots(nrows=2, ncols=3, tight_layout=True)


y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)


x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)
axes[0,1].plot(x1, y1)
axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
              ylim=(-1, 5), yticks=np.arange(-1, 6))


x2 = np.random.normal(15, 5, 2000)
plt.xlim(0, 40)
axes[0,2].hist(x2, bins=10)
axes[0,2].set(yticks=np.array([200, 400]))


# np.random.seed(3)
x4 = 5 + np.random.normal(0, 2, 50)
y4 = 5 + np.random.normal(0, 2, len(x4))
axes[1,0].set(xlim=(-1, 9), xticks=np.array([0, 5]),
              ylim=(-1, 9), yticks=np.array([0.0, 2.5, 5.0, 7.5]))
axes[1,0].scatter(x4, y4)


plt.show()
```
