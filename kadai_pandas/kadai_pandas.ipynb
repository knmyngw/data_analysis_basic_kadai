```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
df = pd.read_csv('sample_pandas_6.csv')
df_category = pd.read_csv('category.csv')
df = df.merge(df_category, on='商品番号', how='inner')
df.head(1)
```

```python
len(df.groupby('カテゴリー'))
```

```python
x = np.arange(len(df.groupby('カテゴリー')))
y = df['カテゴリー'].value_counts()

plt.title('カテゴリー別受注回数')
y.plot(kind='bar')
# y.plot.bar()

plt.show()
```

```python
# 商品番号ごとの注文累計
df.groupby('商品番号')['注文数'].sum()
```

```python
# カテゴリーごとの単価情報
df.groupby('カテゴリー')['単価'].describe()
```
